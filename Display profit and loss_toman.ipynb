```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python

with open('./portfolio_summary.json', 'r', encoding='utf-8') as f:
    data = json.load(f)  

# make DataFrame
df = pd.DataFrame(data)

# Convert datetime to date type
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
# Y-axis formatter to millions
def millions(x, pos):
    return f'{x * 1e-6:.1f} million'
```

```python
## year show 
```

```python
df['year'] = df['datetime'].dt.year

# Sort by time
df_sorted = df.sort_values('datetime')

# Get the latest total_toman value for each year
last_values = df_sorted.groupby('year').tail(1).set_index('year')['total_toman']



# Draw a diagram
fig, ax = plt.subplots(figsize=(12, 4))
last_values.plot(kind='bar', color='tomato', ax=ax)

# Appearance settings
ax.set_title("Amount of assets per year(toman)")
ax.set_ylabel("Amount of assets")
ax.set_xlabel("year")
ax.grid(axis='y')

# Apply formatting to the y-axis
ax.yaxis.set_major_formatter(FuncFormatter(millions))

plt.tight_layout()
plt.savefig('Amount of assets per year(toman)', bbox_inches='tight', dpi=300)
#plt.show()
```

```python
# month show 
```

```python

df['month'] = df['datetime'].dt.to_period('M')  # Monthly period

# Sort by date
df_sorted = df.sort_values('datetime')

# Get the last row of each month
last_monthly_values = df_sorted.groupby('month').tail(1).set_index('month')['total_toman']

# Show or draw
fig, ax = plt.subplots(figsize=(12, 4))
# Draw a diagram
last_monthly_values.plot(kind='bar', figsize=(12, 4), color='slateblue')
ax.set_title("Amount of assets per month (toman)")
ax.set_ylabel("assets per month")
ax.set_xlabel("month")
ax.grid(axis='y')
ax.yaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.savefig('Amount of assets per month (toman)', bbox_inches='tight', dpi=300)
#plt.show()
```

```python
## day
```

```python

df['month'] = df['datetime'].dt.to_period('M')
df['day'] = df['datetime'].dt.to_period('D')

# Default value: the last month in the data
latest_month = df['month'].max()

# 👇 If you want to manually specify the month (for example, May 2025)
#latest_month = pd.Period('2025-12', freq='M')
# Filter data for that month
monthly_data = df[df['month'] == latest_month].sort_values('datetime')

# Get the latest `total_toman` value each day
last_daily_values = monthly_data.groupby('day').tail(1).set_index('day')['total_toman']


fig, ax = plt.subplots(figsize=(12, 4))
last_daily_values.plot(kind='bar', figsize=(12, 4), color='seagreen')
ax.set_title(f"Amount of assets per day(month {latest_month})")
ax.set_ylabel("assets per day")
ax.set_xlabel("day")
plt.grid(axis='y')
ax.yaxis.set_major_formatter(FuncFormatter(millions))

plt.tight_layout()
plt.savefig('Amount of assets per day(toman)', bbox_inches='tight', dpi=300)
#plt.show()
```

```python


def millions(x, pos):
    return f'{int(x/1e6)}M'

try:
    df['datetime'] = pd.to_datetime(df['datetime'])


    df['day'] = df['datetime'].dt.to_period('D')

# Sort by date
    df_sorted = df.sort_values('datetime')
    df_sorted = df_sorted[df_sorted['datetime'] >= df_sorted['datetime'].max() - pd.Timedelta(days=30)]

# Get the latest `total_toman` value for each day
    last_daily_values = df_sorted.groupby('day').tail(1).set_index('day')['total_toman']

# Calculate daily net change
    daily_profit_loss = last_daily_values.diff().dropna()

# Color scheme: green for profit, red for loss
    colors = ['green' if val >= 0 else 'red' for val in daily_profit_loss]


    fig, ax = plt.subplots(figsize=(14, 5))
    daily_profit_loss.plot(kind='bar', color=colors, ax=ax)
   # Draw a diagram
    ax.set_title("Daily net profit/loss (toman)")
    ax.set_ylabel("Change from the previous day")
    ax.set_xlabel("day")
    ax.axhline(0, color='black', linestyle='--')
    plt.xticks(rotation=90)
    plt.grid(axis='y')
    ax.yaxis.set_major_formatter(FuncFormatter(millions))
    plt.tight_layout()
    plt.savefig("daily_profit_loss.png", bbox_inches='tight', dpi=300)
    # plt.show()

except Exception as e:
    print("An error occurred, continuing.:", e)
```
